**ZAD 1**
**Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:**


**• wypełnianie deklaracji PIT,**
- wymaga człowieka, ale bot może pomóc w wypełnieniu na podstawie podanych informacji.

**• streszczanie tekstu,**
- nie wymaga dużej inteligencji, boty radzą sobie z tym dobrze, chociaż mogą wymagać korekty.

**• tłumaczenie tekstu,**
- nie wymaga, bot bez problemu się tym zajmie, można ewentualnie korygować powstałe błędy

**• klasyfikacja tekstu do kategorii tematycznych,**
- raczej nie wymaga

**• odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim),**
- bot może mieć problem np. pytając go o opinię na jakiś temat

**• układanie rozkładu jazdy transportu miejskiego,**
- po wyposażeniu SI w odpowiednie materiały myślę, że mógłby układać rozkłady

**• programowanie (pisanie programów komputerowych),**
- bot się tym zajmie bez problemu

**• „programowanie” kanałów telewizyjnych,**
- Nie wiem. Wydaje mi się, że może do jakiegoś stopnia tylko pomóc w tym

**• testowanie oprogramowania,**
- pewnie zależy jakiego, ale może mieć problem z rozumieniem poszczególnych kontekstów

**• komponowanie muzyki,**
- bot może to robić, ale bedzie słychać raczej że nie było to wyprodukowane przez człowieka. Brakuje elementu emocjonalnego i logiki czasem.

**• rozwiązywanie układów równań,**
- nie wymaga ludzkiej inteligencji, AI radzi sobie z tym

**• symboliczne obliczanie pochodnych funkcji,**
- pewnie AI też sobie poradzi z tym

**• symboliczne całkowanie funkcji,**
- nie wiem

**• kierowanie samochodem.**
- wymaga ludzkiej inteligencji, ze względu na nieprzewidywalność sytuacji na drodze.





**ZAD 2** **Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej inteligencji:**


**• streszczanie tekstu,**
- SI może analizować tekst i tworzyć jego podsumowanie.

**• tłumaczenie tekstu,**
- SI potrafi przetłumaczyć tekst z jednego języka na drugi.

**• klasyfikacja tekstu do kategorii tematycznych,**
- SI może przypisać tekst do odpowiedniej kategorii tematycznej.

**• odpowiadanie na proste pytania zadawane w języku naturalnym,**
- SI najczęściej rozumie takie pytania

**• rozwiązywanie układów równań,**
- SI może je rozwiązywać

**• układanie rozkładu jazdy,**
- SI może w pewnym stopniu analizować dane i tworzyć optymalne rozkłady jazdy, wydaje mi się jednak, że będą one wymagały dużej korekty

**• rozwiązywanie układów równań liniowych,**
- SI może rozwiązywać takie równania

**• symboliczne obliczanie pochodnych,**
- SI może wykonywać operacje matematyczne, takie jak obliczanie pochodnych

**• symboliczne całkowanie,**
- tak samo jak w powyższym punkcie

**• kierowanie samochodem**
- SI może być wykorzystywana w samochodach autonomicznych, aby sterować pojazdem.


**ZAD 3** **Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):**


**• rozmowa towarzyska,**
- SI potrafi prowadzić rozmowy na tematy ogólne, udawać zainteresowanie i odpowiadać w sposób naturalny.


**• dyskusja polityczna,**
- SI może analizować dane i generować argumenty na różne tematy. Jednak nie posiada własnych przekonań i wartości, co może być zauważalne w dyskusji.

**• dyskusja naukowa,**
- SI posiada informacje naukowe i może generować odpowiedzi na pytania, ale nie ma własnej wiedzy i nie potrafi tworzyć nowych teorii.

**• odpowiadanie na pytania klientów w telefonicznej infolinii,**
- może być wykorzystywana w automatycznych systemach telefonicznych, aby odpowiadać na typowe/powtarzające się pytania klientów.

**• odpowiadanie na pytania klientów w internetowej infolinii.**
- jest szeroko wykorzystywana w chatbotach, które odpowiadają na pytania klientów na stronach internetowych.
